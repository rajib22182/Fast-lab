```python
'''10.9 Matplotlib Histograms
In Matplotlib, we use the hist() function to create histograms.
The hist() function will use an array of numbers to create a histogram, the array is sent into the function
as an argument.
'''
import matplotlib.pyplot as plt
import numpy as np

'''
x = np.random.normal(170, 10, 250): Generates 250 random numbers that follow a normal (Gaussian) distribution:

170 is the mean (average) of the distribution.
10 is the standard deviation, which controls the spread of the data around the mean.
250 is the number of random data points to generate.

plt.hist(x): Creates a histogram of the data in x. A histogram is a graphical representation of
the distribution of the data, where the data is divided into bins and the frequency (or count) of data points in 
each bin is plotted.
'''

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('gettyimages-10190248-612x612.jpg', cv2.IMREAD_GRAYSCALE)

# Compute the histogram of the image
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# Plot the histogram
plt.plot(histogram)
plt.title('Histogram of Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()

```
